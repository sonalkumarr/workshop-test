```python
# This is a simple demo code block.

def add_numbers(a, b):
    """This function adds two numbers and returns the result."""
    result = a + b
    return result

# Let's try calling the function with some values
num1 = 15
num2 = 10
sum_result = add_numbers(num1, num2)

print(f"The sum of {num1} and {num2} is: {sum_result}")
```
